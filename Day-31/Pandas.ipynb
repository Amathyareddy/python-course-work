```python
import pandas as pd

prices=[2999, 15999, 52999, 4999, 1999]
products=["wireless Earbuds","Smartphone","Laptop","Smartwatch","Bluetooth Speaker"]

product_prices = pd.Series(prices,index=products)
print(product_prices)
```

```python
print("Mean:", product_prices.mean())
```

```python
print("Sum:", product_prices.sum())
```

```python
print("Max:", product_prices.max())
```

```python
print("Head (Frist 3 Elements):\n", product_prices.head(3))
```

```python
print("Tail (Last 2 Elements):\n", product_prices.tail(2))
```

```python
print("Apply (Adding 18% GST):\n", product_prices.apply(lambda x: f'₹{x+(x*0.18)}'))
```

```python
print("Map (Formatting as Curreny):\n", product_prices.map(lambda x: f"₹{x}.00"))
```

```python
print(product_prices.sort_values())
```

```python
print(product_prices.sort_index())
```

```python
print(product_prices.sort_index(ascending=False))
```

```python
print("Value Counts:\n", product_prices.value_counts())
```

```python
import pandas as pd

data={
       "Product": ["Wireless Earbuds","Smartphone","Laptop","Smartwatch","Bluetooth Speaker"],
       "Brand": ["SoundMax","TechNova","ByteCore","TimeTrack","EchoBoom"],
       "Price": [2999, 15999, 52999,4999, 5678],
       "Stock": [50, 30, 20, 40, 60],
       "BestSeller": [True,False,False,True,True]
       }
df = pd.DataFrame(data)
print(df)
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print(df.head())
```

```python
print(df.head(3))
print(df.tail(2))
```

```python
print(df)
print(df.loc[1, "Product"])
print(df.loc[4, "Product"])
print(df.loc[4, "Stock"])

print(df.iloc[4, 0])
print(df.iloc[2, 1])
print(df.iloc[3, 3])
```

```python
print(df.iloc[3])
print(df.iloc[2])
```

```python
print(df)
print("*************Using loc********************\n", df.loc[df['BestSeller']==True])
print("*************Using loc********************\n", df.loc[df['Stock']<40])
print("*************Using loc********************\n", df.loc[df['Price']<10000])
```

```python
df_renamed = df.rename(columns={"Price": "Cost"})
print("After Renaming 'Price' to 'Cost':\n", df_renamed)
#df.rename(columns={"Product": "Product name"}, inplace=True)
```

```python
df_renamed = df.rename(columns={"Price": "Cost"})
print(df_renamed)
```

```python
df_grouped = df.groupby("Brand").agg({"Price": "sum", "Stock": "sum"})
print(df_grouped)
```

```python
grouped=df.groupby("Brand").agg({"Price": ["mean", "max", "min"]})
print(grouped)
```

```python
grouped = df.groupby("Brand")["Price"].mean()
print(grouped)
```

```python
data2 = {
    "Brand": ["SoundMax", "TechNova", "ByteCore", "TimeTrack", "EchoBoom"],
    "Rating": [4.2, 4.5, 4.0, 4.1, 3.9],
    "discount": [28, 40, 16, 10, 5]
}
df_ratings = pd.DataFrame(data2)
print(df_ratings)
```

```python
df_merged = pd.merge(df, df_ratings, on="Brand")
print("Merged DataFrame (Adding Ratings):\n", df_merged)
```

```python
new_data = {
    "Product": ["Tablet"],
    "Brand": ["SmartWare"],
    "Price": [12999],
    "Stock": [25],
    "BestSeller": [True]
}
df_new = pd.DataFrame(new_data)
print(df_new)
```

```python
df_concat = pd.concat([df, df_new], ignore_index=True)
print("Concatenated DataFrame:\n", df_concat)
```
